### SMOTE


데이타 불균형을 해결하기 위해 언더 샘플링을 진행합니다.


```python
normal_ratio = 1.0  # 1.0 means 1:1 ratio

df_normal = train_data[train_data["target"] == 1]
df_abnormal = train_data[train_data["target"] == 0]

num_normal = len(df_normal)
num_abnormal = len(df_abnormal)
print(f"Original  Total: Normal: {num_normal}, AbNormal: {num_abnormal}")
```

```python
# 원본 데이터프레임을 결합
df_combined = pd.concat([df_normal, df_abnormal], axis=0).reset_index(drop=True)

# 특성과 레이블 분리
X = df_combined.drop(columns=["target"])
y = df_combined["target"]
```

```python
# SMOTE 인스턴스 생성
smote = SMOTE(sampling_strategy='auto', random_state=RANDOM_STATE)

# SMOTE를 사용하여 오버샘플링
X_resampled, y_resampled = smote.fit_resample(X, y)

# 결과 확인
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled["target"] = y_resampled
```

```python
num_normal_resampled = df_resampled[df_resampled["target"] == 0].shape[0]
num_abnormal_resampled = df_resampled[df_resampled["target"] == 1].shape[0]
print(f"Resampled Total: Normal: {num_normal_resampled}, AbNormal: {num_abnormal_resampled}")
```

### 데이터 분할


```python
df_train, df_test = train_test_split(
    df_resampled,
    test_size=0.2,
    stratify=df_resampled["target"],
    random_state=RANDOM_STATE,
)


def print_stats(df: pd.DataFrame):
    num_normal = len(df[df["target"] == 1])
    num_abnormal = len(df[df["target"] == 0])
    print(f"  Total: Normal: {num_normal}, AbNormal: {num_abnormal}" + f" ratio: {num_abnormal/num_normal}")


# Print statistics
print(f"  \tAbnormal\tNormal")
print_stats(df_train)
print_stats(df_test)
```

```python
X_train = df_train.drop(['target'], axis = 1)
y_train = df_train['target']

X_test = df_test.drop(['target'], axis = 1)
y_test = df_test['target']
```
